```python
import pandas as pd
import numpy as np
import joblib
df_test = pd.read_csv('./feature.csv')
x_test = df_test[['Z1','Z2','r1','r2','R1','R2','N1','N2','EP1','EP2','EA1','EA2','IE1','IE2']]
clf = joblib.load('../gbr.pkl')
y_test=clf.predict(x_test)

y_test=pd.DataFrame(y_test)
index = df_test['dacs']

df_pred = './df_pred_go.csv'
pd.concat([index,y_test,x_test], axis=1).to_csv(df_pred, index=False, encoding='utf-8')

df_pred = pd.read_csv(df_pred, sep=',', encoding='utf-8')
df_pred
```
